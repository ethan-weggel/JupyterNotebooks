
## Qestion 3 [20 marks]:
Continue with the 'inflammation-data' directory. Given the following list of file names:
```
filenames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv'],
```
write a program to plot the daily averages, maximums, and minimums of inflammation for each of the file. Your plot should have three rows each of which shows side-by-side plots of average, max, and min. The following figure illustrates the expected outcome:
![](https://i.imgur.com/XXT2Jqw.png)

````python
import os #importing os to join paths cleanly
````

````python
staticDirPath = "/content/drive/MyDrive/Colab Notebooks/Info212/Week 2/inflammation-data" #declaring variable to path of inflammation data directory
fileNames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv'] #declaring the file names I need to iterate through in a list
fig, axes = plt.subplots(3, 3, figsize=(10, 10)) #making the figure and axis' to make all subplots on a single image, 3 rows, 3 columns of subplots

row = 0 #making row variable to keep track of which subplot row we are on
column = 0 #making column variable to keep track of which subplot column we are on

for fileName in fileNames: #iterating through all files in the list of files to plot
  currentData = np.loadtxt(os.path.join(staticDirPath, fileName), delimiter=",") #loading the file into a numpy array
  dailyAverages = currentData.mean(axis=0) #getting the daily averages
  dailyMaximums = currentData.max(axis=0) #getting the daily maximums
  dailyMinimums = currentData.min(axis=0) #getting the dailiy minimums

  axes[row, column].set_title(f"{fileName}") #setting the title of each row as the file name
  axes[row, column].plot(dailyAverages) #plot averages
  axes[row, column].set_ylabel("average") #set the axis name
  column += 1 #incrementing column to move over

  axes[row, column].plot(dailyMaximums) #plotting maximums
  axes[row, column].set_ylabel("max") #set the axis name
  column += 1 #incrememnting column to move over

  axes[row, column].plot(dailyMinimums) #plotting minimums
  axes[row,column].set_ylabel("mix") #set the axis name
  column = 0 #reset the column back to zero for next file's data

  row += 1 #increment row so we can move to a new row for a new file

plt.tight_layout() #adjusting the spacing so there is no overlap
plt.show() #showing the final graphs



````
